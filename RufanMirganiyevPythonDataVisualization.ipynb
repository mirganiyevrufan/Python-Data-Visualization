```python
# 5. Check for outliers using a box plot or histogram.

# We are going to use a new seaborn built-in dataset called healthexp. 
# Basically dataset contains data which country pays more health insurance by year

#loading dataset
healthexp = sns.load_dataset("healthexp")

#tail(30) will display the last 30 rows from the dataset which years are between 2016-2020
healthexp = healthexp.tail(30)
healthexp
```

```python
# sns.boxplot() shows the distribution of quantitative data in a way that 
# facilitates comparisons between variables 'Year' and 'Spending_USD'
box_plot_healthexp = sns.boxplot(x="Year", y="Spending_USD", data=healthexp)
```

```python
#A commonly used rule says that a data point is an outlier if it is more than
#  1.5*IQR(interquartile range) above the third quartile or below the first quartile.
#defining box_plot_outliers function
def box_plot_outliers(series):
    #calculating Q1 ->first quartile
    first_quartile = series.quantile(0.25) #The quantile() method calculates the quantile of the values in a given axis.The default is row axis.
    #calculating Q3 -> third quartile
    third_quartile = series.quantile(0.75)
    #calculating IQR ->interquartile range
    interquartile_range = third_quartile - first_quartile
    # identifiying below Q1 and above Q3
    low_high_interquartile_range = (series < first_quartile - 1.5 * interquartile_range) | (series > third_quartile + 1.5 * interquartile_range)
    series = series[low_high_interquartile_range]
    return series
```

```python
# creates a new pandas dataframe
outliers = pd.DataFrame([])
```

```python
# for loop iterates through unique years
for i in healthexp.copy().Year.unique():
    # creates a copy of healthexp dataset sorted by 'Year'
    outlier = healthexp.copy().sort_values(['Year'])
    # filtering the dataset and resetting the indexes
    outlier = outlier.loc[outlier['Year'] == i, :].reset_index(drop=True)
    # setting index 'Year' and applies 'box_plot_outliers' function
    outlier = outlier[['Year', 'Spending_USD']].set_index('Year').apply(box_plot_outliers)
    #resetting the indexes
    outlier = outlier.reset_index()
    # concatenating outliers with outlier
    outliers = pd.concat([outliers,outlier])
```

```python
#these are the outlier values
outliers.reset_index(drop=True)
```

```python
# in this dataset we are going to calculate fare amount (total amount without tolls, fees)
#and  as well as average speed for whole rides
# genfromtxt() function is very helpful when we are expecting some missing values in the dataset to be loaded
#loading dataset 
nyc_taxi = np.genfromtxt('nyc_taxis.csv',delimiter=',',skip_header=True)

# speed is equal to  trip_distance / trip_length 
# trip_distance is miles so we need to divide trip_length by 3600 to gain speed in miles.
def calc_speed(trip_distance,trip_length):
    try:
        speed = nyc_taxi[:,trip_distance] / (nyc_taxi[:,trip_length] / 3600)
    except:
        print("Please pass specified column numbers only as an argument")
    return speed.mean() # mean() function can be used to calculate mean/average of a given list of numbers. 

# We caluclate the fare_amount by subtracting all fees from total_amount 

def fare_amount(total_amount,fees_amount,tolls_amount,tip_amount):
    try:
        fare_amount =  nyc_taxi[:,total_amount] - nyc_taxi[:,fees_amount] - nyc_taxi[:,tolls_amount] - nyc_taxi[:,tip_amount]
    except:
        print("Please pass specified column numbers only as an argument")
    return fare_amount
    
    
    
    
```
